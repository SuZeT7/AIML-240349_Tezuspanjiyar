```python
# cost func, gradient func, accuracy func
def cost(ytru, ypred):
    a = (ytru - ypred) ** 2
    return a;
print (cost(9,7))
```

```python
def gradient(ytru, ypred):
    a = 2 * (ypred - ytru)
    return a;
print (gradient(9,7))
```

```python
def accu(ytru,ypred):
    if (ytru == ypred):
        return 1;
    else:
        return 0;
print (accu(9,7))
```
